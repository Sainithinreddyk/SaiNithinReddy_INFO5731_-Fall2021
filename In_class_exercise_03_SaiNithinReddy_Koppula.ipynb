```python
# You code here (Please add comments in the code):

#Intially I have loaded all the necessary libraries
from bs4 import BeautifulSoup
import requests
import urllib.request
import re
import pandas as pd

```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.amazon.com/product-reviews/B0B9HGV6JB/ref=acr_dp_hist_5?ie=UTF8&filterByStar=five_star&reviewerType=all_reviews#reviews-filter-bar'

headers = {'User-Agent': 'Mozilla/5.0'}

titles = []
text = []

for pg in range(1, 150):
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    titles += [title.get_text().strip() for title in soup.find_all('a', class_='review-title-content')]
    text += [review.get_text().strip() for review in soup.find_all('span', class_='review-text-content')]


```

```python
# Tile column contains the title and reciew contains the text
data = {'Title': titles, 'Review': text}
raw_data = pd.DataFrame(data)
raw_data.info()
```

```python
raw_data.head()
```

**SENTENCE COUNT**

```python
#Counting the number of sentences in each review
raw_data['Number  of sentences'] = raw_data['Review'].apply(lambda y: len(str(y).split(".")))
raw_data.head()
```

**WORD COUNT**

```python
#Countong the number of words in a review
raw_data['Number of Words'] = raw_data['Review'].apply(lambda y: len(str(y).split(" ")))
raw_data.head()
```

**CHARECTER COUNT**

```python
#Counting the number of charecters in each review
raw_data['Number of Charecters'] = raw_data['Review'].str.len()
raw_data.head()
```

**SPECIAL CHARECTERS COUNT**

```python
#Counting the number of special charecters
def spchar(x):
    red = 0
    for i in range(len(x)):
        if not((x[i].isalpha()) and (x[i].isdigit())):
            red = red + 1
    return red

raw_data['Number of special charecters'] = raw_data['Review'].apply(lambda y: spchar(y))
raw_data.head()
```

**NUMERIC COUNT**

```python
raw_data['Num of num'] = raw_data['Review'].apply(lambda x: len([x for x in x.split() if x.isdigit()]))
raw_data.head()
```

**REMOVING ALPHANUMERIC**

```python
raw_data['Modified_review'] = raw_data['Review'].apply(lambda x: x.replace('[^\w\s]',''))
raw_data.head()
```

**REMOVING NUMERIC DATA**

```python
raw_data['Modified_review'] = raw_data['Modified_review'].str.replace('\d', '')
raw_data.head()
```

**REMOVING SPECIAL CHARECTERS**

```python

raw_data["Modified_review"] = raw_data['Modified_review'].apply(lambda t : ''.join(re.sub(r"[^a-zA-Z0-9]+", ' ', ch) for ch in t))
raw_data.head()
```

**COVERTING STRING TO LOWER CASE**

```python
raw_data['Modified_review'] = raw_data['Modified_review'].apply(lambda text: " ".join(word.lower() for word in text.split()))
raw_data.head()
```

**REMOVING STOP WORDS**

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = stopwords.words('english')
raw_data['Modified_review'] = raw_data['Modified_review'].apply( lambda text : " ".join(w for w in text.split() if w not in stop_words) )
raw_data.head()
```

**LEMATIZZING**

```python
import nltk
nltk.download('omw-1.4')
from textblob import Word
nltk.download('wordnet')
raw_data['Modified_review'] = raw_data['Modified_review'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
raw_data.head()
```

**POS OF EACH WORD IN REVIEW**

```python
nltk.download('averaged_perceptron_tagger')
from nltk import pos_tag
raw_data['POS_Review'] = raw_data['Modified_review'].str.split().map(pos_tag)
raw_data.head()
```

**OCCURENCE OF POS**

```python
tags = ['JJ', 'RB', 'CC', 'IN', 'UH', 'NN', 'VB', 'PRP', 'PDT', 'DT']
count_cols = ['num of adj', 'num of adverb', 'num of conjunction', 'num of subordinating conjunction',              'num of interjection', 'num of noun', 'num of verb', 'num of pronoun',              'num of predeterminer', 'num of determiner']

def count_tags(tags_list, tagged_text):
    counts = []
    for tag in tags_list:
        tag_count = [1 for word, word_tag in tagged_text if word_tag == tag]
        counts.append(sum(tag_count))
    return counts

raw_data[count_cols] = raw_data['POS_Review'].apply(lambda x: pd.Series(count_tags(tags, x)))
raw_data.head()
```
